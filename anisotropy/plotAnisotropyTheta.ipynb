# Plot Anisotropy #

#### Preamble ####

```python
import numpy as np
import numpy.random as rn
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

from scipy.linalg import expm

matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preamble'] = r'\usepackage{bm}'
matplotlib.rcParams['font.size'] = 9
matplotlib.rcParams['legend.fontsize'] = 8
```

```python
#### Load data ####
data = np.load("anisotropy.npz")
data2 = np.load("anisotropy_theta_phi.npz")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3, 2.5))

ax.set_xlabel("$\\tan{\\theta}/\sqrt{2}$")
# ax.set_xlabel("$\sqrt{\\frac{\lambda_x^2 + \lambda_y^2}{2 \lambda_z^2}}$")
ax.set_ylabel("$\mathcal{E}/\mathcal{E}_{\\rm{max}}$")

##### Data ######
ax.plot(np.tan(data['theta'])/np.sqrt(2), data['entropy1D']/np.log(2), color='g', label='single-bath')
ax.plot(np.tan(data['theta'])/np.sqrt(2), data['entropy3D']/np.log(2), color='C0', label='three-bath')
ax.plot(1, 0.085, 'ko', markersize=4)
ax.plot(1, 0.555, 'ko', markersize=4)
ax.axvline(1, color='k', linestyle='--', lw=0.8, dashes=(5, 4))

plt.legend(frameon=False)
plt.savefig('anisotropy_theta.pdf', bbox_inches="tight", dpi=500)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel("$\\tan{\\theta}/\sqrt{2}$")
ax.set_ylabel("$\\phi$")
ax.set_zlabel("$\mathcal{E}/\mathcal{E}_{\\rm{max}}$")

Y, X = np.meshgrid(np.tan(data2['theta'])/np.sqrt(2), np.tan(data2['phi']))
Z1D = data2['entropy1D']/np.log(2)
Z3D = data2['entropy3D']/np.log(2)

##### Data ######

ax.plot_surface(X, Y, Z1D.reshape(X.shape), cmap='Greens', label='single-bath')
ax.plot_surface(X, Y, Z3D.reshape(X.shape), cmap='Blues', label='three-bath')

ax.plot3D(1, 1, 0.085, 'ko', markersize=4, zorder=4)
ax.plot3D(1, 1, 0.555, 'ko', markersize=4, zorder=4)

plt.savefig('anisotropy_theta_phi.pdf', bbox_inches="tight", dpi=500)
```

```python
fig = plt.figure(figsize=(7.3, 3))

ax = fig.add_subplot(1, 2, 1)

ax.set_xlabel("$\\tan{\\theta}/\sqrt{2}$")
ax.set_ylabel("$\mathcal{E}/\mathcal{E}_{\\rm{max}}$")

##### Data ######
ax.plot(np.tan(data['theta'])/np.sqrt(2), data['entropy1D']/np.log(2), color='g', label='single-bath')
ax.plot(np.tan(data['theta'])/np.sqrt(2), data['entropy3D']/np.log(2), color='C0', label='three-bath')
ax.plot(1, 0.085, 'ko', markersize=4)
ax.plot(1, 0.555, 'ko', markersize=4)
ax.axvline(1, color='k', linestyle='--', lw=0.8, dashes=(5, 4))

ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.set_xlabel("$\\tan{\\theta}/\sqrt{2}$")
ax.set_ylabel("$\\tan{\\phi}$")
ax.set_zlabel("$\mathcal{E}/\mathcal{E}_{\\rm{max}}$")

Y, X = np.meshgrid(np.tan(data2['theta'])/np.sqrt(2), np.tan(data2['phi']))
Z1D = data2['entropy1D']/np.log(2)
Z3D = data2['entropy3D']/np.log(2)

##### Data ######
ax.plot_surface(X, Y, Z1D.reshape(X.shape), cmap='Greens', label='single-bath')
ax.plot_surface(X, Y, Z3D.reshape(X.shape), cmap='Blues', label='three-bath')

ax.plot3D(1, 1, 0.085, 'ko', markersize=4, zorder=4)
ax.plot3D(1, 1, 0.555, 'ko', markersize=4, zorder=4)

plt.savefig('anisotropy.pdf', bbox_inches="tight", dpi=500)
```
