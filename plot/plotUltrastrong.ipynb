```python
data_10 = np.array(pd.read_csv(r'../other_data/sz_alpha_2D_sx_sz_10.csv'))
data_20 = np.array(pd.read_csv(r'../other_data/sz_alpha_2D_sx_sz_20.csv'))
data_30 = np.array(pd.read_csv(r'../other_data/sz_alpha_2D_sx_sz_30.csv'))
data_40 = np.array(pd.read_csv(r'../other_data/sz_alpha_2D_sx_sz_40.csv'))
```
