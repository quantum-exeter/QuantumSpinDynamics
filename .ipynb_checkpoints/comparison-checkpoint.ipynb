## Variables

### Lorentzian Spectral Density

```julia
prm1 = [7.0 5.0 10.0];
prm2 = [1.4 0.5 0.16];
prm3 = [1.4 0.001 10];
```

```julia
œâ0, Œì, Œ± = prm3*œâL;
```

### RC-Specific

```julia
Œ© = œâ0;
Œª = œâL*sqrt(Œ±/Œ©);
Œ¥ = Œì/(2œÄ*Œ©);
n = 10;
```

```julia
Htot(n) = HRC1D(n, Œª, Œ©);
```

```julia
gibbs(Htot(5), 5, 1)
```

### Environment

```julia
T = 200;
```

### Timescales

```julia
ti, tf, dt = 0.0, 20000.0, 100.0;
tspan = (ti, tf);
t = ti:dt:tf;
```

## Initial States

```julia
Œ∏, œï = -œÄ/2, 0; # Define some Œ∏ and œï on the Bloch sphere
```

```julia
state0 = œÅ0(Œ∏, œï, n, Œ©, T);
```

## MFGS

```julia
Htot(n) = HRC1D(n, Œª, Œ©);
```

```julia
œÅtot(n) = gibbs(Htot(n), T);
```

```julia
œÅspMFGS(n) = ptrace(œÅtot(n), n);
```

```julia
ni, nf, dn = 1, 100, 1;
nlist = ni:dn:nf;
```

```julia
expz(n) = real(tr(œÅspMFGS(n)*sz0)); # Sz expectation value
sz = [expz(i) for i in nlist];
dn = DataFrame(hcat(nlist, sz), :auto);
CSV.write("MFGS_convergence200.csv", dn, header = ["n", "sz"]); # Exports RC data to a CSV file
```

## Solving the Master Equation

```julia
superop = ùíÆ(n, Œ©, Œª, Œ¥, T); # This is the superoperator corresponding to the Redfield master equation
```

```julia
alg = Vern7();
```

```julia
dstate(dœÅ, œÅ, v, t) = mul!(dœÅ, superop, œÅ); # Solves the DE
prob = ODEProblem(dstate, vec(state0), tspan);
sol = solve(prob, alg);
solF(t) = reshape(sol(t), (2n, 2n)); # Reformats vector into a density matrix

expx(t) = real(tr(solF(t)*kron(sx0, ùïÄb(n))));  # Sx expectation value
expy(t) = real(tr(solF(t)*kron(sy0, ùïÄb(n)))); # Sy expectation value
expz(t) = real(tr(solF(t)*kron(sz0, ùïÄb(n)))); # Sz expectation value
```

```julia
time = [i for i in t];
sx = [expx(i) for i in t];
sy = [expy(i) for i in t];
sz = [expz(i) for i in t];
```

```julia
df = DataFrame(hcat(time, sx, sy, sz), :auto);
```

```julia
CSV.write("rc_data_prm3_10.csv", df, header = ["t", "sx", "sy", "sz"]); # Exports RC data to a CSV file
```
